```python
import requests
import json
base_url = 'https://data.gov.au/data/api/3/action/datastore_search?resource_id=6ab35f9a-e476-4d29-84de-2e18d1e704c7'
query = 'Eliza'
limit = '10'
response = requests.get(base_url+"&q="+query+"&limit="+limit)
searchresult = json.loads(response.text)

from collections import OrderedDict
convicts = searchresult["result"]["records"]
for c in convicts:
    print(c["Vessel"],"\t",c["DateofDeparture"],"\t",c["PlaceofArrival"])
```

```python
import requests
import json
base_url = 'https://api.github.com/users/postgres/repos'
response = requests.get(base_url)
repos = json.loads(response.text)

for repo in repos:
    print(repo["name"], ":", repo["language"])
    
last_3_repositories = sorted(repos, key = lambda r: r["created_at"], reverse=True)[:3]

for repo in last_3_repositories:
    print(repo["name"], ":", repo["language"])
```

```python
import requests
import json
# Need API to authenticate each request to Google Maps Platform
base_url = 'http://maps.googleapis.com/maps/api/geocode/json'
my_params= {'address': '1 Cleveland Street,Darlington,Australia','language':'en'}
response = requests.get(base_url, params = my_params)
            
print(json.dumps(response.json(), indent=4, sort_keys=False))
            
results  = response.json()['results']
x_geo = results[0]['geometry']['location']
print(x_geo['lng'], x_geo['lat'])
```
